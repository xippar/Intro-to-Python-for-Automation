# `string` ข้อมูลประเภทตัวอักษร

### ข้อมูลที่เป็นตัวอักษร

ข้อมูลประเภทตัวอักษรในการทำงานกับ Python จะถูกจัดเก็บในประเภทข้อมูลที่เรียกว่า string ซึ่งจะเป็นชุดข้อมูลที่มีการเก็บข้อมูลส่วนที่เป็นเนื้อหาของตัวอักษรที่ใช้งาน เช่น ชื่อ ที่อยู่ หรือรายละเอียดอื่นๆ ที่ต้องบันทึกเป็นตัวอักษร

การเก็บข้อมูลเป็นตัวอักษรมาพร้อมกับเครื่องมือ function ที่เน้นการทำงานร่วมกับชุดข้อมูลที่เป็นตัวอักษร เช่น การเปลี่ยนคำให้เป็นตัวพิมพ์ใหญ่ การแบ่งคำ การดึงตัวอักษรจากคำ การแสดงผลข้อมูล เป็นต้น

### String

`string` เป็นประเภทของ `object` ที่เน้นเก็บข้อมูลที่มีคุณสมบัติเป็นตัวอักษร และหลายๆ ครั้งที่ข้อมูลที่ดึงเข้ามาใน Jupyter Notebook จะถูกสร้างเป็นข้อมูลประเภท `string` ก่อน ดังนั้นความเข้าใจข้อมูลประเภทนี้เป็นส่วนสำคัญในการจัดการข้อมูลในภาษา Python

ในขณะที่ string มีความใกล้เคียงตัวอักษรมากที่สุด string ไม่ใช่ตัวอักษรทั้งหมด แต่เป็น Object ที่ประกอบไปด้วย Object ที่เป็นตัวอักษร และ Object ที่เกี่ยวข้องกับการจัดหน้าตาของข้อความ

## 1) สร้าง `string` ผ่าน `' '`

เมื่อมีสัญลักษณ์ `' '` `string` `object` จะถูกสร้างขึ้นมาทันที

```python
''
```

**ข้อสังเกต:** `string` `object` จะมีสีแดง

## 2) เก็บข้อมูลเป็น `string`

ใส่ตัวอักษรใดๆ ลงไประหว่าง `' '` เพื่อเก็บตัวอักษรนั้นเป็นข้อมูลประเภท `string`

```python
'hello world'
```

## 3) ข้อมูลใน `string`

เราจะเรียก `'h'` ว่า `string` `object` **ที่มี 1 `object` อยู่ด้านใน**

เราจะเรียก `'hello world'` ว่า `string` `object` **ที่มี 11 `object` อยู่ด้านใน**

ซึ่งตัวอักษรแต่ละตัว `h` `e` `l` `l` `o` ` ` `w` `o` `r` `l` `d` ล้วนเก็บเป็น `string`

## 4) ดึง `object` ใน `string` ด้วยตำแหน่ง

เราจะใช้สัญลักษณ์ `[ ]` เป็นคำสั่งเพื่อดึง `object` ที่เก็บไว้ใน `string`

โดยระบุ `index` หรือ ตำแหน่ง ที่เริ่มต้นด้วย 0

```python
'hello world'[0]
```
